```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("IRIS.csv")
```

```python
df
```

```python
df.describe()
```

```python
df.info()
```

```python
#grouping by qualitiative variable

df1 = df.groupby('species')
df1.describe()
```

```python
df1.mean()
```

```python
df1.median()
```

```python
df1.std()
```

```python
df1.count()
```

```python
df1.min()
```

```python
df1.max()
```

```python
df1['species'].unique()
```

```python
summary_stats = df1.agg(['mean', 'std', 'median', 'count', 'min', 'max'])

species_list = list(df1.groups)

print(summary_stats)
print(species_list)
```
